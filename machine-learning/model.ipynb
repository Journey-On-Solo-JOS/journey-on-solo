```python
users = 'https://raw.githubusercontent.com/Journey-On-Solo-JOS/journey-on-solo/main/machine-learning/dataset/users.csv'
df_user = pd.read_csv(users)
df_user
```

```python
# Merge users and items dataset on common preference column
user_features = df_user.drop(["user_id"], axis=1)
item_features = df.drop(columns=['rating', 'reviews_count', 'long', 'lat'])
```

```python
# Create interaction matrix
interaction_matrix = np.dot(user_features, item_features.values.T)
interaction_matrix
```

```python
from sklearn.preprocessing import StandardScaler

# Scale user and item features
scaler = StandardScaler()
user_features_scaled = scaler.fit_transform(user_features)
item_features_scaled = scaler.fit_transform(item_features)
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(user_features_scaled, interaction_matrix, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create a recommendation system
user_input = keras.layers.Input(shape=[X_train.shape[1]])
item_input = keras.layers.Input(shape=[X_train.shape[1]])

user_embedding = keras.layers.Dense(64, activation='relu')(user_input)
user_embedding = keras.layers.Dense(32, activation='relu')(user_embedding)

item_embedding = keras.layers.Dense(64, activation='relu')(item_input)
item_embedding = keras.layers.Dense(32, activation='relu')(item_embedding)

concat = keras.layers.Concatenate()([user_embedding, item_embedding])
output = keras.layers.Dense(1)(concat)

model = keras.Model(inputs=[user_input, item_input], outputs=output)
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
history = model.fit([X_train, X_train], y_train, validation_data=([X_test, X_test], y_test), epochs=100, batch_size=64)
```

```python
# Predict preferences for test data
test_user_data = np.repeat(X_test, item_features_scaled.shape[0], axis=0)
test_item_data = np.tile(item_features_scaled, (X_test.shape[0], 1))
predictions = model.predict([test_user_data, test_item_data])

def recommend(user_id, df, predictions):
    # Ensure predictions are 1D and of correct length
    user_predictions = predictions.ravel()[:df.shape[0]]
    df['predictions'] = user_predictions
    return df

# Capture new user input
def capture_new_user_input():
    print("Enter new user preferences:")
    preferences = {}
    for column in user_features.columns:
        preferences[column] = float(input(f"{column}: "))
    return pd.DataFrame([preferences])

# Update user dataset with new input
new_user_data = capture_new_user_input()
df_user = pd.concat([df_user, new_user_data], ignore_index=True)

# Recalculate interaction matrix with updated user data
user_features = df_user.drop(["user_id"], axis=1)
user_features_scaled = scaler.fit_transform(user_features)
interaction_matrix = np.dot(user_features, item_features.values.T)

# Predict preferences for new user data
test_user_data = np.repeat(user_features_scaled[-1].reshape(1, -1), item_features_scaled.shape[0], axis=0)
test_item_data = np.tile(item_features_scaled, (1, 1))
predictions = model.predict([test_user_data, test_item_data])

# Recommend items for the new user
recommended_items = recommend(len(df_user) - 1, df, predictions)
print(recommended_items)
```
